#Questão 1 (vale até 2 pontos)
Determine a área entre as curvas definidas por $x_1=y^2-2$ e $x_2=y$. Trace o gráfico destas funções sombreando em azul claro a área entre elas. Você pode usar apenas o python, matplotlib, numpy e sympy. Neste caderno do Google Colab.

## Resposta

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

#Cáculo da área
y = sp.symbols('y')
func1 = y**2 - 2
func2 = y

relacao = sp.Eq(func1, func2)
interseccao = sp.solve(relacao)

i1 = interseccao[0]
i2 = interseccao[1]

fA = y- (y**2 - 2)

area = sp.integrate(fA, (y, interseccao[0], interseccao[1]))
print(f"Área entre curvas: {area}")

#Plotando gráfico
x = np.linspace(-2, 3)

def x1(y):
  return y**2 - 2

def x2(y):
  return y

plt.plot(x, x1(x))
plt.plot(x, x2(x))
plt.scatter(interseccao[0], x1(interseccao[0]), color="black", s=15)
plt.scatter(interseccao[1], x1(interseccao[1]), color="black", s=15)
plt.fill_between(x, x1(x), x2(x), alpha=0.3, where=(x > interseccao[0]) & (x < interseccao[1]))
plt.grid()
plt.show()
```

Considere a função $f(x)$ dada por:

$$f(x) = 3x^2 - 4x + 2$$

Calcule a antiderivada $F(x)$ da função $f(x)$. Utilize o Teorema Fundamental do Cálculo para encontrar o valor exato da integral definida de $f(x)$ no intervalo $[1, 3]$.

Mostre todos os passos de cálculo, usando a notação matemática (latex) neste caderno do Google Colab e justifique suas respostas utilizando as propriedades das operações integração.

## Resposta



Para achar a antiderivada de $f(x)$ precisamos integrar a função:
$$\int_{1}^{3} f(x)$$
$$\int_{1}^{3} 3x^2 - 4x + 2$$
Agora usamos a regra da soma ou diferença:
$$\int_{1}^{3} 3x^2 \int_{1}^{3}- 4x \int_{1}^{3} 2$$
Agora calculamos as integrais definidas de cada uma das integrais definidas:
$$\int 3x^2 = \frac{3x^3}{3} = x^3$$
$$\int- 4x = -\int 4x = - \frac{4x^2}{2} = -2x^2$$
$$\int 2 = 2x$$
Agora podemos aplicar a regra fundamental do para cada derivada definida:
$$\int_{1}^{3} 3x^2 = (3)^3 - (1)^3 = 26$$
$$\int_{1}^{3} -4x = -2(3)^2 - (-2(1)^2) = -16$$
$$\int_{1}^{3} 2 = 2(3) - 2(1) = 4$$
Somando tudo obtemos o resultado:
$$F(x) = 26 -16 + 4 = 14$$
